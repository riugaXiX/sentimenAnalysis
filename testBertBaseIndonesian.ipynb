```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import pandas as pd
```

```python
df = pd.read_csv('dataset/data_clean')
df.drop(['Value'], axis=1,inplace=True)
df['stemmed'] = df['stemmed'].str.slice(0,512)
```

```python
model_name = "cahya/bert-base-indonesian-522M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Buat list kosong untuk menyimpan hasil sentimen, label, dan score
sentiments = []
labels = []
scores = []

# Loop melalui setiap teks dalam dataset
for text in df['stemmed']:
    # Tokenisasi teks dan konversi ke tensor
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)

    # Mengambil hasil sentimen, label, dan score
    sentiment_label = torch.argmax(outputs.logits, dim=1).item()
    sentiment_score = torch.softmax(outputs.logits, dim=1)[0][sentiment_label].item()

    # Mapping label sentimen ke teks
    if sentiment_label == 1:
        sentiment_text = "Positif"
    else:
        sentiment_text = "Negatif"

    # Simpan hasil sentimen, label, dan score ke dalam list
    sentiments.append(sentiment_text)
    labels.append(sentiment_label)
    scores.append(sentiment_score)

# Tambahkan kolom baru ke dalam DataFrame untuk menyimpan hasil sentimen, label, dan score
df['Sentiment'] = sentiments
df['Label'] = labels
df['Score'] = scores
```
