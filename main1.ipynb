# Cleaning Data

### note : menghapus bahasa asing pada dataset 

```python
def clean_data(sentence):
    sentence = sentence.translate(str.maketrans(' ',' ',string.punctuation))
    sentence = re.sub('<.*>',' ',sentence)
    sentence = re.sub('[^a-zA-Z]',' ', sentence)
    sentence = re.sub("\n"," ", sentence)
    sentence = re.sub(r"\b[a-aZ-Z]\b"," ", sentence)
    # sentence = ' '.join([w for w in sentence.split() if w != text])
    sentence = ' '.join(sentence.split())
    sentence = sentence.lower()
    return sentence

def remove_dash(sentence):
    parts = sentence.split(" - ", 1)
    if len(parts) > 1:
        sentence = parts[1]
    else:
        sentence = sentence
    return sentence

def remove_minus(sentence):
    parts = sentence.split(" – ", 1)
    if len(parts) > 1:
        sentence = parts[1]
    else:
        sentence = sentence
    return sentence

# def remove_spesifikword(sentence):
#     sentence = sentence.str.replace()
```

```python
df['Konten'] = [remove_dash(i) for i in konten]
konten = df['Konten']
```

```python
df['Konten'] = [remove_minus(i) for i in konten]
konten = df['Konten']
```

```python
df['Konten'] = [clean_data(i) for i in konten]
konten = df['Konten']
```

```python
df.dropna(inplace=True, axis=1)
```

# Tokenizing

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'\w+|$[0-9]+|\S')
df["Token"] = df["Konten"].apply(regexp.tokenize)
```

# Removing Stop Word

```python
import nltk as nk
nk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopword = stopwords.words('indonesian')
txt_stopword = pd.read_csv('stopword.txt', names=['stopword'], header=None)
stopword.extend(['wkwk','hahahaha','haha','yabg'])
stopword.extend(txt_stopword["stopword"][0].split('\n'))
stopword = set(stopword)
```

```python
def stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['stopword'] = df['Token'].apply(stopwords)

df['stopword']
```

# Filtering

```python
txt_spesifik_kata = pd.read_csv("spesifikkata", names=["kata"])
spesifikkata = [i for i in txt_spesifik_kata['kata']]
```

```python
def removing_Specific_word(text):
    text = [word for word in text if word not in spesifikkata]
    return text
```

```python
df['stopword'] = df['stopword'].apply(removing_Specific_word)
```

# Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

kata = df['stopword']
for doc in kata:
    for term in doc:
        if term not in term_dict:
            term_dict[term] = ''
    
    
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
#     print(term,":",term_dict[term])
#     print(len(term_dict))
#     print('--------------------------------')
```

```python
def get_stemmed_term(doc):
    return [term_dict[term] for term in doc]

df['stemmed'] = df['stopword'].apply(get_stemmed_term)
df['stemmed']
```
